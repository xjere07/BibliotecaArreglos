```python
# 1. Dibuje el patrón de una flecha utilizando for:

c = 5
centro = c

for i in range(1, c + 1):
    for j in range(1, 2 * c):
        if (i == c) or (j == centro) or (j == centro - (i - 1) and i > 1) or (j == centro + (i - 1) and i > 1):
            print("*", end="")
        else:
            print(" ", end="")
    print()

for i in range(1, c + 1):
    for j in range(1, 2 * c):
        if j == centro:
            print("*", end="")
        else:
            print(" ", end="")
    print()

print()
```

```python
# 2. Realice un algoritmo que cuente el número de palabras de una cadena de texto.
def contar_palabras(texto):
    palabras = texto.split()
    return len(palabras)

frase = input("Ingrese una cadena de texto: ")
cantidad = contar_palabras(frase)
print("El nUmero de palabras es: ", cantidad)
```

```python
# 3. Cree un algoritmo que permita saber si dos frases son anagramas.
def son_anagramas(frase1, frase2):
  # .lower me ayuda hacer minusculas
    f1 = frase1.lower()
    f2 = frase2.lower()

    texto1 = ""
    texto2 = ""
    for caracter in f1:
        if caracter != " ":
            texto1 = texto1 + caracter

    for caracter in f2:
        if caracter != " ":
            texto2 = texto2 + caracter

    if len(texto1) != len(texto2):
        return False

    lista1 = list(texto1)
    lista2 = list(texto2)
  # .sort nos ayuda a ordenar los elementos
    lista1.sort()
    lista2.sort()

    return lista1 == lista2

frase1 = input("Ingrese la primera frase: ")
frase2 = input("Ingrese la segunda frase: ")

if son_anagramas(frase1, frase2):
    print("Las frases SI son anagramas")
else:
    print("Las frases NO son anagramas")
```

```python
# 4. Determinar cuántos dígitos tiene un número entero.
def contar_digitos(numero):
    contador = 0

    while numero > 0:
        numero = numero // 10
        contador = contador + 1
    return contador


num = int(input("Ingrese un numero entero: "))

if num > 0:
    cantidad = contar_digitos(num)
    print("El numero tiene", cantidad, "digitos")
else:
    print("Por favor ingrese un numero mayor que 0")
```

```python
# 5. Implemente el algoritmo de ordenamiento por selección.
def ordenamiento_seleccion(lista):
    cantidad = len(lista)

    for i in range(cantidad):
        posicion_minimo = i

        for j in range(i + 1, cantidad):
            if lista[j] < lista[posicion_minimo]:
                posicion_minimo = j

        valor_auxiliar = lista[i]
        lista[i] = lista[posicion_minimo]
        lista[posicion_minimo] = valor_auxiliar

    return lista

numeros = [10, 25, 8, 2, 11]

print("Lista original:", numeros)
lista_ordenada = ordenamiento_seleccion(numeros)
print("Lista ordenada:", lista_ordenada)
```
